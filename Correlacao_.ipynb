```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

c = np.array([ 5.6, 5.2, 4.8, 4.5, 4.4, 2.9, 2.7 ])

idade = np.array([ 1, 2, 3, 4, 5, 6, 7 ])

dados = pd.DataFrame({
    'Conversao alimentar' :
    c,
    'Idade' : idade
})

dados.head() #mostra o DataFrame

dados.corr() #Correlação 

dados['Conversao alimentar'].mean() #média 

dados['Conversao alimentar'].std() #desvio padrão amostral

np.std(dados['Conversao alimentar'], ddof=1) #desvio padrão populacional (ddof 0 - Populacional, ddof 1- Amostral)

plt.plot(dados['Idade'], dados['Conversao alimentar'])

plt.title('Tabela do Boi Gordo')
```
