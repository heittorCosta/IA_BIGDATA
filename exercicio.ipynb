```python
import pandas as pd
import numpy as np

dados = pd.read_json(path_or_buf='imoveis.json', orient='columns')
dados.ident[0]
dados.listing[0]
```

```python
dados_lista1 = pd.json_normalize(dados.ident)
dados_lista1.head()
```

```python
dados_lista2 = pd.json_normalize(dados.listing)
dados_lista2.head()
```

```python
dados_moveis = pd.concat([dados_lista1, dados_lista2], axis= 1)
dados_moveis.head()
```
