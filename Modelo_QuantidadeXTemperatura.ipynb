```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

#Dados de treinamento
quantidade = np.array([100, 200, 300, 400, 500, 600, 700]).reshape(-1, 1)
temperatura = np.array([30, 35, 40, 45, 60, 100, 150]).reshape(-1, 1)
#Criar e treinar o modelo de regressão linear
modelo = LinearRegression()
modelo.fit(quantidade, temperatura)

quantidade_teste = np.array([100, 200, 70, 300, 500, 1000]).reshape(-1, 1)
temperatura_previsto = modelo.predict(quantidade_teste)

for i in range(len(quantidade_teste)):
    print(f"Quantidade: {quantidade_teste[i][0]} Jaules - Temperatura prevista: {temperatura_previsto[i]} Cº")
plt.scatter(quantidade, temperatura, color = 'blue', label = 'Dados de Treinamento')
plt.plot(quantidade_teste, temperatura_previsto, color = 'red', label = 'Dados de teste')
plt.xlabel('Quantidade)')
plt.ylabel('Temperatra (Cº)')
plt.title('Relação entre Quantidade e Temperatura')
plt.legend()
plt.grid(True)
plt.show()
```
