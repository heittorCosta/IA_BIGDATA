```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

#Dados de treinamento
area = np.array([50, 100, 120, 300]).reshape(-1, 1)
valor = np.array([180.000, 300.000, 375.000, 600.000]).reshape(-1, 1)
#Criar e treinar o modelo de regressão linear
modelo = LinearRegression()
modelo.fit(area, valor)

area_teste = np.array([100, 200, 70, 300, 500, 1000]).reshape(-1, 1)
valor_previsto = modelo.predict(area_teste)

for i in range(len(area_teste)):
    print(f"Area: {area_teste[i][0]} m² - Valor previsto: R$ {valor_previsto[i]} ")
plt.scatter(area, valor, color = 'blue', label = 'Dados de Treinamento')
plt.plot(area_teste, valor_previsto, color = 'red', label = 'Dados de teste')
plt.xlabel('Area')
plt.ylabel('Valor (R$)')
plt.title('Relação entre Area e Valor de um Imóvel')
plt.legend()
plt.grid(True)
plt.show()
```
