```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dados_x = np.linspace(0, 10, 10)
dados_y = dados_x * dados_x

plt.figure( figsize=( 10, 5 ), facecolor= 'yellow' )

ax = plt.axes()
ax.set_facecolor('white')

plt.plot(dados_x, dados_y, color = 'red', markersize = 5, linestyle = 'dashed', markeredgecolor = 'black', linewidth = 2, marker = 'o')
plt.title('Grfico Simples', color = 'blue', fontfamily = 'comic sans ms')
plt.xlabel('Eixo X', color = 'blue', fontweight = 'bold', fontfamily = 'comic sans ms')
plt.ylabel('Eixo Y', color = 'blue' , fontweight = 'bold', fontfamily = 'comic sans ms')
plt.rcParams.update({'font.size' : 15})
plt.figure(figsize=( 10, 10 ))
linhas = 2
colunas =2
plt.subplot(linhas, colunas, 1)
plt.plot(dados_x, 'r--')
plt.title('Grfico 1')

#Grfico 2
plt.subplot(linhas, colunas, 2)
plt.plot(dados_y, 'b--')
plt.title('Grfico 2')

#Grfico3
plt.subplot(linhas, colunas, 3)
plt.plot(dados_y, 'y--')
plt.title('Grfico 3')

#Grfico4
plt.subplot(linhas, colunas, 4)
plt.plot(dados_y * 3, 'g--')
plt.title('Grfico 4')

plt.savefig('4Graficos')
```
