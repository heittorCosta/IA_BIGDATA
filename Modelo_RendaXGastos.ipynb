```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

#Dados de treinamento
renda = np.array([50, 100, 120, 300]).reshape(-1, 1)
gastos = np.array([180.000, 300.000, 375.000, 600.000]).reshape(-1, 1)
#Criar e treinar o modelo de regressão linear
modelo = LinearRegression()
modelo.fit(renda, gastos)

renda_teste = np.array([100, 200, 70, 300, 500, 1000]).reshape(-1, 1)
gastos_previsto = modelo.predict(renda_teste)

for i in range(len(renda_teste)):
    print(f"Renda: {renda_teste[i][0]}  - Gastos previsto: R$ {gastos_previsto[i]} ")
plt.scatter(renda, gastos, color = 'blue', label = 'Dados de Treinamento')
plt.plot(renda_teste, gastos_previsto, color = 'red', label = 'Dados de teste')
plt.xlabel('Renda')
plt.ylabel('Gastos (R$)')
plt.title('Relação entre renda e Gastos')
plt.legend()
plt.grid(True)
plt.show()
```
